```python
import pandas as pd
import pyodbc

try:
    # üîπ Conexi√≥n a SQL Server (autenticaci√≥n de Windows)
    conexion = pyodbc.connect(
        "Driver={ODBC Driver 17 for SQL Server};"
        "Server=localhost\\SQLEXPRESS;"
        "Database=POCO;"
        "Trusted_Connection=yes;"
    )

    print("‚úÖ Conexi√≥n a SQL Server exitosa.")

    # üîπ Consulta SQL
    SQL_QUERY = "SELECT * FROM Sales"

    # üîπ Ejecutar y guardar resultados en un DataFrame
    df_sales = pd.read_sql(SQL_QUERY, conexion)
    print(f"üìä DataFrame creado exitosamente con {len(df_sales)} filas.")
    print(df_sales.head())  # muestra las primeras filas

except Exception as e:
    print(f"‚ùå Error al conectar o ejecutar la consulta: {e}")

finally:
    # üîπ Cerrar conexi√≥n correctamente
    if 'conexion' in locals():
        conexion.close()
        print("üîå Conexi√≥n a SQL Server¬†cerrada.")

# 4. Mostrar el DataFrame
print("\nPrimeras 5 filas del DataFrame:")
print(df_sales.head())
print("\nInformaci√≥n del DataFrame:")
df_sales.info()
```
